### 1.	Write a Pandas program to compare the elements of two Pandas series. Sample Series [4,65,436,3,9], [7,0,3,897,9]


```python
import pandas as pd

ser_1 = pd.Series([4, 65, 436, 3, 9])
ser_2 = pd.Series([7, 0, 3, 897, 9])

result = ser_1 == ser_2
print(result)
```

### 2.	Write a Pandas program to add, subtract, multiply and divide two Pandas series. Sample Series: [2,4,6,8,14], [1,3,5,7,9]


```python
series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])

# Addition
addition_result = series1 + series2
print("Addition Result:")
print(addition_result)

# Subtraction
subtraction_result = series1 - series2
print("\nSubtraction Result:")
print(subtraction_result)

# Multiplication
multiplication_result = series1 * series2
print("\nMultiplication Result:")
print(multiplication_result)

# Division
division_result = series1 / series2
print("\nDivision Result:")
print(division_result)
```

### 5.	Write a Pandas program to display the most frequent value in the given series and replace everything else as ‘Other’ in the series. (Use the ‘HomeTeamGoals’ column in the dataset provided)

```python
df=pd.read_csv('C:\\Users\\USER\\Desktop\\y2 Recess\\BSSE 2301 EXERCISE 2\\AfricaCupofNationsMatches.csv')
df
```

```python
df.info()
```

```python
# Get the most frequent value
most_frequent_value = df['HomeTeamGoals'].mode().iloc[0]
print(most_frequent_value)

```

```python
# Replace everything else as 'Other'
df['HomeTeamGoals'] = df['HomeTeamGoals'].fillna('Other').map({most_frequent_value: most_frequent_value, 'Other': 'Other'})

print(df['HomeTeamGoals'])
```
